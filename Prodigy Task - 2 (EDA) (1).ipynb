```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('E:/Titanic.csv')
print(df)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='copper')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Distribution of Survival')
plt.show()
```

```python
df.groupby(['Sex', 'Survived'])['Survived'].count()
```

```python
df[['Sex','Survived']].groupby(['Sex']).mean().plot.bar()
sns.countplot('Sex',hue='Survived',data=df,)
plt.show()
```

```python
pd.crosstab([df.Sex,df.Survived],df.Pclass,margins=True).style.background_gradient(cmap='summer_r')
```

```python
sns.factorplot('Pclass', 'Survived', hue='Sex', data=df)
plt.show()
```

```python
print('Oldest person Survived was of:',df['Age'].max())
print('Youngest person Survived was of:',df['Age'].min())
print('Average person Survived was of:',df['Age'].mean())
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
sns.violinplot('Pclass','Age',hue='Survived',data=df,split=True,ax=ax[0])
ax[0].set_title('PClass and Age vs Survived')
ax[0].set_yticks(range(0,110,10))
sns.violinplot("Sex","Age", hue="Survived", data=df,split=True,ax=ax[1])
ax[1].set_title('Sex and Age vs Survived')
ax[1].set_yticks(range(0,110,10))
plt.show()
```
